- use the output of 5.1 to turn LGA based data into electorate based using proportions

- note: users will have to manually uncomment some code to merge desired files

```python
import pandas as pd
```

```python
age = pd.read_csv('../data/curated/sudoData/age.csv')
edu = pd.read_csv('../data/curated/sudoData/edu.csv')
income = pd.read_csv('../data/curated/sudoData/income.csv')
population = pd.read_csv('../data/curated/sudoData/population.csv')
```

```python
Electorate_LGA_conversion = pd.read_csv('../data/curated/GeographicalProportionateData/Elect_LGA.csv')
```

```python
Electorate_LGA_conversion = Electorate_LGA_conversion[Electorate_LGA_conversion['percentage_overlap'] != 0]
```

```python
# merge them
# merged = Electorate_LGA_conversion.merge(age, how = 'left', left_on = 'lga_code17', right_on = 'lga_code_2016')
# merged = Electorate_LGA_conversion.merge(edu, how = 'left', left_on = 'lga_code17', right_on = 'lga_code_2016')
merged = Electorate_LGA_conversion.merge(income, how = 'left', left_on = 'lga_code17', right_on = 'lga_code_2021')
# merged = Electorate_LGA_conversion.merge(population, how = 'left', left_on = 'lga_code17', right_on = 'lga_code_2021')
merged
```

```python
merged = merged.dropna()
```

```python
# merged['avg_age_voters'] = merged['avg_age_voters'].fillna(0)
# merged['avg_year_completeld'] = merged['avg_year_completeld'].fillna(0)
merged['avg_weekly_income'] = merged['avg_weekly_income'].fillna(0)
# merged['erp_2021'] = merged['erp_2021'].fillna(0)
# merged['net_migration_2020_21'] = merged['net_migration_2020_21'].fillna(0)
```

```python
# create the output by iterating through each LGA and adding together proportions

Elect_div = list()
value_list1 = list()
value_list2 = list()
for id, data in merged.groupby(['Elect_div']):
    
    # data['proportionated'] = data['avg_age_voters']*data['percentage_overlap']
    # data['proportionated'] = data['avg_year_completeld']*data['percentage_overlap']
    data['proportionated'] = data['avg_weekly_income']*data['percentage_overlap']
    # data['proportionated'] = data['erp_2021']*data['percentage_overlap']
    # data['proportionated2'] = data['net_migration_2020_21']*data['percentage_overlap']
    
    Elect_div.append(id)

    # because use 2017n LGA but census data might be based on other years
    total = sum(data['percentage_overlap'])

    data['proportionated'] = data['proportionated']/total
    # data['proportionated2'] = data['proportionated2']/total

    value_list1.append(sum(data['proportionated']))
    # value_list2.append(sum(data['proportionated2']))
```

```python
# join the individual columns up and output
# pd.DataFrame({'DivisionName': Elect_div, 'avg_age': value_list1}).to_csv('../data/curated/sudoData/age_proportionated.csv')
# pd.DataFrame({'DivisionName': Elect_div, 'avg_educ_years': value_list1}).to_csv('../data/curated/sudoData/edu_proportionated.csv')
pd.DataFrame({'DivisionName': Elect_div, 'avg_weekly_income': value_list1}).to_csv('../data/curated/sudoData/income_proportionated.csv')
# pd.DataFrame({'DivisionName': Elect_div, 'erp': value_list1, 'net_migration': value_list2}).to_csv('../data/curated/sudoData/population_proportionated.csv')
```
